```python
'''
准备数据阶段。
'''
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

train = pd.read_csv('../datasets/fashionmnist/fashion-mnist_train.csv')

train, dev = train_test_split(train, test_size=10000, random_state=2019)

test = pd.read_csv('../datasets/fashionmnist/fashion-mnist_test.csv')

y_train = train['label'].values
y_dev = dev['label'].values
y_test = test['label'].values

X_train = train.drop('label', axis=1)
X_dev = dev.drop('label', axis=1)
X_test = test.drop('label', axis=1)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_dev = ss.transform(X_dev)
X_test = ss.transform(X_test)

train_samples, n_features = np.shape(X_train)
dev_samples = len(X_dev)
test_samples = len(X_test)

print train_samples, dev_samples, test_samples

'''
构建卷积神经网络阶段（Construction Phase)。
'''
import tensorflow as tf

batch_size = 1024
n_batches = train_samples / batch_size
n_epochs = 50
learning_rate = 1e-3


X = tf.placeholder(tf.float32, [None, 784])
y = tf.placeholder(tf.int32)

x_images = tf.reshape(X, [-1, 28, 28, 1])

conv_1 = tf.layers.conv2d(inputs=x_images, filters=16, kernel_size=3, strides=[1, 1], padding='SAME', activation=tf.nn.tanh, name='C1')

pool_1 = tf.layers.average_pooling2d(inputs=conv_1, pool_size=2, strides=[2,2], padding='SAME', name='P1')

conv_2 = tf.layers.conv2d(inputs=pool_1, filters=32, kernel_size=3, strides=[1, 1], padding='SAME', activation=tf.nn.tanh, name='C2')

pool_2 = tf.layers.average_pooling2d(inputs=conv_2, pool_size=2, strides=[2,2], padding='SAME', name='P2')

conv_3 = tf.layers.conv2d(inputs=pool_2, filters=64, kernel_size=3, strides=[1, 1], padding='SAME', activation=tf.nn.tanh, name='C3')

conv_3_flat = tf.reshape(conv_3, [-1, 64 * 7 * 7])

fc_1 = tf.layers.dense(inputs=conv_3_flat, units=100, activation=tf.nn.tanh, name='FC_1')

logits = tf.layers.dense(inputs=fc_1, units=10, activation=None, name='FC_2')

xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)

loss = tf.reduce_mean(xentropy, name='loss')

optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)

training_op = optimizer.minimize(loss)

correct = tf.nn.in_top_k(logits, y, 1)

accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

'''
执行训练网络阶段（Executation Phase)。
'''
saver = tf.train.Saver()

init = tf.global_variables_initializer()

with tf.Session() as sess:
    init.run()
    max_dev_acc = -np.inf
    
    for iteration in range(n_epochs):
        for i in range(n_batches):
            np.random.seed(iteration * n_batches + i)
            indices = np.random.randint(train_samples, size=batch_size)
            X_batch = X_train[indices]
            y_batch = y_train[indices]
            sess.run(training_op, feed_dict={X: X_batch, y: y_batch})
        
        dev_acc = accuracy.eval(feed_dict = {X: X_dev, y: y_dev})
        
        print 'Dev_Accuracy @ Iteration %d: %f' % (iteration + 1, dev_acc)
        if max_dev_acc < dev_acc:
            saver.save(sess, '../models/section_4.1.9/baseline_cnn_model.ckpt')
            print '#### Save model @ Iteration %d ####' %(iteration + 1)
            max_dev_acc = dev_acc
        
with tf.Session() as sess:
    # 加载所有相关参数。
    saver.restore(sess, '../models/section_4.1.9/baseline_cnn_model.ckpt')
    print '验证集上的分类准确率为：%f' % accuracy.eval(feed_dict = {X: X_dev, y: y_dev})
    print '测试集上的分类准确率为：%f' % accuracy.eval(feed_dict = {X: X_test, y: y_test})
```

```python
'''
准备数据阶段。
'''
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

train = pd.read_csv('../datasets/fashionmnist/fashion-mnist_train.csv')

train, dev = train_test_split(train, test_size=10000, random_state=2019)

test = pd.read_csv('../datasets/fashionmnist/fashion-mnist_test.csv')

y_train = train['label'].values
y_dev = dev['label'].values
y_test = test['label'].values

X_train = train.drop('label', axis=1)
X_dev = dev.drop('label', axis=1)
X_test = test.drop('label', axis=1)

ss = StandardScaler()
X_train = ss.fit_transform(X_train)
X_dev = ss.transform(X_dev)
X_test = ss.transform(X_test)

train_samples, n_features = np.shape(X_train)
dev_samples = len(X_dev)
test_samples = len(X_test)

print train_samples, dev_samples, test_samples

'''
构建卷积神经网络阶段（Construction Phase)。
'''
import tensorflow as tf

batch_size = 1024
n_batches = train_samples / batch_size
n_epochs = 50
learning_rate = 1e-3


X = tf.placeholder(tf.float32, [None, 784])
y = tf.placeholder(tf.int32)
is_training = tf.placeholder(tf.bool)

x_images = tf.reshape(X, [-1, 28, 28, 1])

conv_1 = tf.layers.conv2d(inputs=x_images, filters=16, kernel_size=3, strides=[1, 1], padding='SAME', activation=None, name='C1')
# conv_1后面，激活函数前面增加BN。
conv_1_bn = tf.layers.batch_normalization(conv_1, training=is_training)
conv_1_bn_out = tf.nn.tanh(conv_1_bn)

pool_1 = tf.layers.average_pooling2d(inputs=conv_1_bn_out, pool_size=2, strides=[2,2], padding='SAME', name='P1')

conv_2 = tf.layers.conv2d(inputs=pool_1, filters=32, kernel_size=3, strides=[1, 1], padding='SAME', activation=None, name='C2')
# conv_2后面，激活函数前面增加BN。
conv_2_bn = tf.layers.batch_normalization(conv_2, training=is_training)
conv_2_bn_out = tf.nn.tanh(conv_2_bn)

pool_2 = tf.layers.average_pooling2d(inputs=conv_2_bn_out, pool_size=2, strides=[2,2], padding='SAME', name='P2')

conv_3 = tf.layers.conv2d(inputs=pool_2, filters=64, kernel_size=3, strides=[1, 1], padding='SAME', activation=tf.tanh, name='C3')

conv_3_flat = tf.reshape(conv_3, [-1, 64 * 7 * 7])

fc_1 = tf.layers.dense(inputs=conv_3_flat, units=100, activation=tf.nn.tanh, name='FC_1')

logits = tf.layers.dense(inputs=fc_1, units=10, activation=None, name='FC_2')

xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)

loss = tf.reduce_mean(xentropy, name='loss')

optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate)

update_ops = tf.get_collection(tf.GraphKeys.UPDATE_OPS)

with tf.control_dependencies(update_ops):
    training_op = optimizer.minimize(loss)

correct = tf.nn.in_top_k(logits, y, 1)

accuracy = tf.reduce_mean(tf.cast(correct, tf.float32))

'''
执行训练网络阶段（Executation Phase)。
'''
saver = tf.train.Saver()

init = tf.global_variables_initializer()


with tf.Session() as sess:
    init.run()
    max_dev_acc = -np.inf
    
    for iteration in range(n_epochs):
        for i in range(n_batches):
            np.random.seed(iteration * n_batches + i)
            indices = np.random.randint(train_samples, size=batch_size)
            X_batch = X_train[indices]
            y_batch = y_train[indices]

            sess.run(training_op, feed_dict={X: X_batch, y: y_batch, is_training: True})
        
        dev_acc = accuracy.eval(feed_dict = {X: X_dev, y: y_dev, is_training: False})
        
        print 'Dev_Accuracy @ Iteration %d: %f' % (iteration + 1, dev_acc)
        if max_dev_acc < dev_acc:
            saver.save(sess, '../models/section_4.1.9/bn_cnn_model.ckpt')
            print '#### Save model @ Iteration %d ####' %(iteration + 1)
            max_dev_acc = dev_acc
        
with tf.Session() as sess:
    # 加载所有相关参数。
    saver.restore(sess, '../models/section_4.1.9/bn_cnn_model.ckpt')
    print '验证集上的分类准确率为：%f' % accuracy.eval(feed_dict = {X: X_dev, y: y_dev, is_training: False})
    print '测试集上的分类准确率为：%f' % accuracy.eval(feed_dict = {X: X_test, y: y_test, is_training: False})
```
