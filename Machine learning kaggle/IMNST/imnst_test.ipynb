```python
import pandas as pd

train_data = pd.read_csv('datasets/train.csv',dtype='int32')
test_data = pd.read_csv('datasets/test.csv',dtype='int32')
x_train = train_data.iloc[:,1:]
y_train = train_data.iloc[:,0]
x_test = test_data
```

```python
print(train_data.shape)
print(test_data.shape)
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(x_train.dtypes)
```

```python
# https://docs.w3cub.com/tensorflow~python/tf/contrib/learn/linearclassifier/
# https://tensorflow.juejin.im/tutorials/kernel_methods.html
import tensorflow as tf             # 1.0版本的tensorflow才有contrib.learn
feature_columns = [tf.contrib.layers.real_valued_column("", dimension=784)]
optimizer = tf.train.FtrlOptimizer(learning_rate=0.01, l2_regularization_strength=1.0)
classifier = tf.contrib.learn.LinearClassifier(feature_columns=feature_columns, n_classes=10,optimizer=optimizer)
classifier.fit(x_train,y_train,batch_size=200,max_steps=4000)
report = classifier.evaluate(x_train,y_train)
print('report:',report)
linear_y_predict = classifier.predict(x_test)
linear_submission = pd.DataFrame({'ImageId':range(1,28001),'Label':linear_y_predict})
linear_submission.to_csv('datasets/linear_submission.csv',index=False)
```

```python
import tensorflow as tf             # 1.0版本的tensorflow才有contrib.learn
feature_columns = [tf.contrib.layers.real_valued_column("", dimension=784)]
# optimizer = tf.train.FtrlOptimizer(learning_rate=0.01, l2_regularization_strength=1.0)
dnn_classifier = tf.contrib.learn.DNNClassifier(feature_columns=feature_columns,hidden_units=[200,50, 10], n_classes=10,model_dir="/models/dnn_classifier")
dnn_classifier.fit(x_train,y_train,batch_size=50,max_steps=5000)
report = dnn_classifier.evaluate(x_train,y_train)
print('dnn_report:',report)
dnn_y_predict = dnn_classifier.predict(x_test)
dnn_submission = pd.DataFrame({'ImageId':range(1,28001),'Label':dnn_y_predict})
dnn_submission.to_csv('datasets/dnn_submission.csv',index=False)
```
